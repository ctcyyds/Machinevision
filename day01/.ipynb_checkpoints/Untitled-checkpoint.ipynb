```python
import cv2 as cv

# 读取图片 参数1：图片路径， 参数2：读取的方式 
img = cv.imread("img/lena.jpg",cv.IMREAD_COLOR)
# 显示窗口 参数1：窗口名称， 参数2：图片数据
cv.imshow("src",img)

# 让程序处于等待推出状态
cv.waitKey(0)
# 当程序推出时，释放所有窗口资源
cv.destroyAllWindows()
```

### 显示一张图片
如果是显示一张图片的话只需要执行cv.imshow&cv.waitKey即可
但是在jupyter中还需要

```python
import cv2 as cv
# 显示方式1
# 导入图片
img = cv.imread("img/lena.jpg")
# 显示图片
cv.imshow("图片展示",img)
cv.waitKey(1000)
```

```python
# 显示方式2
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread("img/lena.jpg")
plt.imshow(img)
```
