# 线性回归

```python
import numpy as np
data = np.array([[80,200],
                 [95,230],
                 [104,245],
                 [112,247],
                 [125,259],
                 [135,262]])
```

```python
import matplotlib.pyplot as plt
for x,y in data:
    plt.scatter(x,y,c="red")
plt.show()
```

 假设房屋价格与面积无关
 - y = b

```python
# 准备数据
b_range = np.linspace(100,400,50) # 100~400取50个点
b_range
```

```python
# 真实值（源于真实数据）
real_price = data[:,1]
# 计算mse mse = (预测值 - 真实值)**2
for b in b_range:
    mse = np.sum((b - real_price)**2)
    # 绘图
    plt.scatter(b,mse,c="red")
plt.show()
```

##梯度下降

```python
# 初始化b
b = 1
# 学习速率
learning_rate = 0.001
# 梯度下降函数
def gradient_decent():
    global b,learning_rate
    # 均方误差
    mse = 0
    # 导函数运算结果
    b_slop = 0
    for real in data[:,1]:
        mse += (b - real)**2
        # 计算导函数
        b_slop += 2 * (b - real)
    # 更新b值   
    b = b - b_slop * learning_rate
    print(f"mse={mse},b_slop={b_slop},b={b}")
```

```python
gradient_decent()
```

```python
gradient_decent()
```

```python
for i in range(25):
    gradient_decent()
```

```python
## 计算y = mx + b
# 初始化m,b
m = 1
b = 1
# 初始化学习速率
learning_rate = 0.00001
def gradient_decent2():
    global m,b,learning_rate
    mse = 0
    m_slop = 0
    b_slop = 0
    for x,real in data:
        # 计算mse
        mse += (m*x + b - real)**2
        # 计算m的偏导数
        m_slop += (m*x + b - real)*x
        # 计算b的偏导数
        b_slop += (m*x + b - real)
    # 更新m,b的值
    m = m - m_slop*learning_rate
    b = b - b_slop*learning_rate
    print(f"mse={mse},m={m},b={b}")
    return m,b,mse
```

```python
m,b,mse = 0,0,0
```

```python
# 机器学习训练
for i in range(29):
        gradient_decent2()
```
